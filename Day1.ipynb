Different Types of Printing Basics in Python

```python
a = int(input("Value 1 :"))
b = int(input("Value 2 :"))
c = a + b
#Method 1
print ("Value 1 is",a)
print ("Value 2 is",b)
#Method 2
print ("Value 1 is",a,"\nValue 2 is",b)
#Method 3
print (f"Value 1 is {a} Value 2 is {b} Value 3 is {c}")
#Method 4
print ("Value 1 is {} Value 2 is {} Value 3 is {}".format(a,b,c))
```
